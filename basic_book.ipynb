```python
# Class

class Calc: # 클래스
    def __init__(self, a): # 메소드
        self.a = a     
    def add(self, b): # 메소드
        print(self.a + b)        
    def multiply(self, b): # 메소드
        print(self.a * b)

calc = Calc(3) # Calc 클래스의 인스턴스 생성
calc.add(4) # add 메소드 실행
calc.multiply(4) # multiply 메소드 실행

class CalcPlus(Calc): # Calc 클래스를 상속받음
    def substract(self, b):
        print(self.a - b)
    def divide(self, b):
        print(self.a / b)

calc_plus = CalcPlus(3)
calc_plus.add(4) # Calc 클래스의 메소드 사용 가능
calc_plus.multiply(4) # Calc 클래스의 메소드 사용 가능
calc_plus.substract(4)
calc_plus.divide(4)
```


### 미분


#### 상미분(Ordinary differential)
- $x$ 값이 변할 때 $\Delta x$에 대한 $f(x)$의 변화율
\
$\cfrac{f(x + \Delta x) - f(x)}{\Delta x}$
- $\Delta x$값을 $0$에 한없이 가까워지게 만들면, 새로운 함수 $f'(x)$를 얻을 수 있음
\
$ f'(x) = \lim\limits_{\Delta x \to 0}\cfrac{f(x + \Delta x) - f(x)}{\Delta x}$


- $f'(x)$는 $f(x)$의 도함수


- $f'(x)$를 구하는 과정을 '함수$f(x)$를 $x$에 관해 미분한다'고 말함


- $f'(x)$의 동일한 식
\
$f'(x)= \cfrac{d}{dx}f(x)$ $\qquad (\cfrac{d}{dx}$ : $x$로 미분)


- 함수의 변수가 $x$뿐인 경우, 하나의 변수에 대한 미분을 상미분이라고 함


- $x$의 변화에 따른 $f(x)$의 변화율 : 기울기


- 도함수를 통해 하나의 변수에 대해 한 점에서의 기울기를 구할 수 있음(접선의 기울기)
\
함수 $f(x)$ 곡선 위의 접점 $(a, f(a))$에서의 기울기는 $f'(a)$

#### 미분법의 기본 공식

- 몇 가지 함수에서 미분 공식을 이용하면 도함수를 간단하게 구할 수 있음


- $f(x) = x^r$(r은 실수)에 대한 도함수
\
$\cfrac{d}{dx}f(x) = rx^{r-1}$


- 두 함수의 합 $f(x)+g(x)$의 미분
\
$\cfrac{d}{dx}(f(x)+g(x))=\cfrac{d}{dx}f(x)+\cfrac{d}{dx}g(x)$


- 두 함수의 곱 $f(x)g(x)$의 미분
\
$\cfrac{d}{dx}(f(x)g(x))=f(x)\cfrac{d}{dx}g(x)+g(x)\cfrac{d}{dx}f(x)$


- 상수는 미분 공식 밖으로 꺼낼 수 있음
\
$\cfrac{d}{dx}kf(x) = k\cfrac{d}{dx}f(x)$


- 지수함수 미분
\
$\cfrac{d}{dx}a^x = a^xlog a$


- 자연지수함수 미분
\
$\cfrac{d}{dx}e^x = e^x$


- 자연로그함수 미분
\
$\cfrac{d}{dx}log x = \cfrac{1}{x}$

#### 연쇄법칙(Chain rule)
- 역전파에 활용됨


- 합성함수(여러 개의 함수를 합성하여 나타낸 함수)
\
$f(g(x)) \to$ $\begin{align} & y=f(u) \\ & u=g(x)\end{align}$


- 예를 들어, 함수 $y=(x^2+1)^3$은 $x^2+1$를 $u$로 치환한 합성함수로 생각할 수 있음
\
$y=u^3$, $u=x^2+1$


- 합성함수의 미분은 식을 구성하는 각 함수에 대한 도함수의 곱으로 계산(연쇄법칙)
\
$\cfrac{dy}{dx} = \cfrac{dy}{du}\cfrac{du}{dx}$


- $\begin{align} & y=f(u) \\ & u=g(x)\end{align}$일 때 $y$를 $x$로 미분하는 도함수의 연쇄법칙 증명
\
$\begin{align}\cfrac{dy}{dx} &= \lim\limits_{\Delta x \to 0} \cfrac{f(g(x+\Delta x)) - f(g(x))}{\Delta x} \\ &= \lim\limits_{\Delta x \to 0} ( \cfrac{f(g(x+\Delta x)) - f(g(x))}{g(x+\Delta x) - g(x)} \cdot \cfrac{g(x+\Delta x) - g(x)}{\Delta x} )\end{align}$
\
\
$\Delta u = g(x + \Delta x) - g(x)$로 두면, $\Delta x \to 0, \Delta u \to 0$ 이므로 연쇄법칙이 유도됨
\
$\begin{align} \cfrac{dy}{dx} &= \lim\limits_{\Delta x \to 0} ( \cfrac{f(u+\Delta u) - f(u)}{\Delta u} \cdot \cfrac{\Delta u}{\Delta x} ) \\ &= \lim\limits_{\Delta x \to 0} ( \cfrac{f(u+\Delta u) - f(u)}{\Delta u} \cdot \lim\limits_{\Delta x \to 0} \cfrac{\Delta u}{\Delta x}) \\ &= \cfrac{dy}{du}\cfrac{du}{dx} \end{align}$

#### 편미분(Partial differentiation)
- 역전파에 활용됨


- 여러 개의 변수를 가진 함수를 하나의 변수로만 미분하는 것


- 미분하는 변수 외의 변수는 상수로 취급


- $\partial$
\
라운드 디, 파셜 디, 파셜


- 편미분 식
\
$\cfrac{\partial}{\partial x} f(x,y) = \lim\limits_{\Delta x \to 0} \cfrac{f(x+\Delta x,y) - f(x,y)}{\Delta x}$


- $\Delta x$를 무한소로 변화시켜 $x$가 한없이 0에 가까워지게 함, $x$에 대해 편미분할 때 $y$는 상수로 취급함


- 예를 들어, $f(x,y) = 3x^2 + 4xy + 5y^3$를 $x$에 대해 편미분
\
$\cfrac{\partial}{\partial x} f(x,y) = 6x + 4y$


- 편미분으로 구한 함수 : 편도함수


- $x$로 편미분
\
$y$값을 고정시켰을 때 $x$의 변화에 따른 $f(x,y)$의 변화율


#### 전미분(Total differential)
- 여러 변수의 변화에 따른 함수의 변화율


- $z = f(x,y)$
\
$\begin{align} \Delta z &= f(x + \Delta x,y + \Delta y) - f(x,y) \\ &= f(x + \Delta x,y + \Delta y) -f(x,y+\Delta y) + f(x,y + \Delta y) - f(x,y) \\ &= \cfrac{f(x + \Delta x,y + \Delta y) -f(x,y+\Delta y)}{\Delta x} \Delta x + \cfrac{f(x,y + \Delta y) -f(x,y)}{\Delta y}\Delta y \end{align}$
\
\
$\Delta x $와 $\Delta y$를 0에 한없이 가까워지게 하여 무한소에 수렴시킴
\
$\lim\limits_{\Delta x \to 0 \Delta y \to 0}\Delta z = \lim\limits_{\Delta x \to 0 \Delta y \to 0} \cfrac{f(x + \Delta x,y + \Delta y) -f(x,y+\Delta y)}{\Delta x} \Delta x + \lim\limits_{\Delta x \to 0 \Delta y \to 0}\cfrac{f(x,y + \Delta y) -f(x,y)}{\Delta y}\Delta y$
\
\
$\Delta x $와 $\Delta y$가 충분히 작아지면 우변 1항의 y는 무시할 수 있음
\
\
편미분 공식을 적용해 좌변을 $dz$로 두고, $\Delta x $와 $\Delta y$를 $dx$, $dy$로 두면 다음 식이 유도됨
\
$dz = \cfrac{\partial z}{\partial x}dx + \cfrac{\partial z}{\partial y}dy$
\
\
$dz$를 $z=f(x,y)$의 전미분이라고 함
\
\
2개 이상의 다변수 함수가 있으므로, 전미분을 일반적인 형태의 공식으로 표현함
\
$dz = \sum\limits_{i} \cfrac{\partial z}{\partial x_i}d x_i$

#### 다변수의 연쇄법칙
- $\cfrac{dz}{dx}$를 구하는 방법 $\qquad$ $\left( \begin{align} & z=f(u,v) \\ & u=g(x) \\ & v=h(x) \end{align} \right)$
\
$dz = \cfrac{\partial z}{\partial u} du + \cfrac{\partial z}{\partial v} dv$
\
$\cfrac{dz}{dx} = \cfrac{\partial z}{\partial u} \cfrac{du}{dx} + \cfrac{\partial z}{\partial v} \cfrac{dv}{dx}$
\
일반 공식 : $\cfrac{dz}{dx} = \sum\limits_{i} \cfrac{\partial z}{\partial u_i} \cfrac{d u_i}{dx}$
\
\
$u_i$는 위 식에서 $u$와 $v$에 들어가는 함수, 동일한 과정을 합성함수에 적용 $ \qquad \left( \begin{align} & z=f(u,v) \\ & u=g(x,y) \\ & v=h(x,y) \end{align} \right)$
\
\
$z$의 $x$에 대한 변화율(편미분) : $\cfrac{\partial z}{\partial x} = \cfrac{\partial z}{\partial u}\cfrac{\partial u}{\partial x} + \cfrac{\partial z}{\partial v} \cfrac{\partial v}{\partial x}$ 
\
\
$z$의 $y$에 대한 변화율(편미분) : $\cfrac{\partial z}{\partial y} = \cfrac{\partial z}{\partial u}\cfrac{\partial u}{\partial y} + \cfrac{\partial z}{\partial v} \cfrac{\partial v}{\partial y}$
\
\
$x_i$가 함수 $z$를 구성하는 하나의 변수일 때, 식 일반화 : $\cfrac{\partial z}{\partial x_i} = \sum\limits_{j} \cfrac{\partial z}{\partial u_j} \cfrac{\partial u_j}{\partial x_i}$
\
\
위 식을 벡터와 행렬의 곱으로 나타냄($z$가 변수$x_1,x_2,\cdots,x_n$의 함수, 그 사이에 들어가는 함수의 개수 $m$)
\
\
$\begin{align} ( \cfrac{\partial z}{\partial  x_1},\cfrac{\partial z}{\partial x_2}, \cdots, \cfrac{\partial z}{\partial x_n} ) &= (\cfrac{\partial z}{\partial u_1},\cfrac{\partial z}{\partial u_2}, \cdots, \cfrac{\partial z}{\partial u_m}) \begin{pmatrix} \cfrac{\partial u_1}{\partial x_1} & \cfrac{\partial u_1}{\partial x_2} & \cdots & \cfrac{\partial u_1}{\partial x_n} \\ \cfrac{\partial u_2}{\partial x_1} & \cfrac{\partial u_2}{\partial x_2} & \cdots & \cfrac{\partial u_2}{\partial x_n} \\ \vdots & \vdots & \ddots & \vdots \\ \cfrac{\partial u_m}{\partial x_1} & \cfrac{\partial u_m}{\partial x_2} & \cdots & \cfrac{\partial u_m}{\partial x_n} \end{pmatrix} \\ &= (\sum\limits_{k=1}^m \cfrac{\partial z}{\partial u_k}\cfrac{\partial u_k}{\partial x_1}, \sum\limits_{k=1}^m \cfrac{\partial z}{\partial u_k}\cfrac{\partial u_k}{\partial x_2}, \cdots, \sum\limits_{k=1}^m \cfrac{\partial z}{\partial u_k}\cfrac{\partial u_k}{\partial x_n}) \end{align}$
\
\
다변수의 연쇄법칙이 일반화된 공식으로 표현됨
\
\
신경망은 다변수의 합성함수로 생각할 수 있기 떄문에, 다변수의 연쇄법칙을 사용해 각 파라미터가 전체에 미치는 영향을 구할 수 있음
\
\
행렬의 곱셈을 이용하면 모든 변수에 대한 편도함수를 한번에 구할 수 있음


### 정규분포


#### 정규분포(Normal distribution)

- 정규분포는 가우스 분포라고도 부르며, 자연계에서 인간 행동, 성질 등 다양한 현상에 잘 들어맞는 데이터 분포



- 신경망에는 다양하게 변화하는 파라미터가 있는데, 그런 파라미터들의 초기값은 정규분포에 따라 무작위로 설정함


- 제품 크기, 사람 키, 시험 성적 등은 대체로 정규분포를 따름

- 가로축은 특정 값, 세로축은 그 값의 빈도와 확률 표시


- $\mu$ : 평균, $\sigma$ : 표준편차(데이터가 평균을 중심으로 얼마나 퍼져 있는지를 표현하는 척도)


- 공식
\
$\mu = \cfrac{\sum\limits_{k=1}^n x_k}{n}$
\
$\sigma = \sqrt{\cfrac{\sum\limits_{k=1}^n (x_k - \mu)^2}{n}}$

```python
# 평균 및 표준편차
import numpy as np

a = np.array([1,2,3,4,5])

print("평균 : ", np.average(a))
print("표준편차 : ", np.std(a))
```

#### 확률밀도함수
- 정규분포는 확률밀도함수로 나타냄


- $y=\cfrac{1}{\sqrt{2\pi\sigma^2}}exp(-\cfrac{(x-\mu)^2}{2\sigma^2})$


- $\mu = 0,\sigma = 1$ 이라면,
\
$y=\cfrac{1}{\sqrt{2\pi}}exp(-\cfrac{x^2}{2})$

```python
# 정규분포 히스토그램
import numpy as np
import matplotlib.pyplot as plt

x = np.random.normal(50, 10, 10000) # 평균 50, 표준편차 10, 10000개 난수 생성

plt.hist(x, bins=50) # bins : 막대 수

plt.show()

print(np.average(x))
print(np.std(x))
```
